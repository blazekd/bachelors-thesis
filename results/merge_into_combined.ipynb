```python
import pandas as pd
import os
import glob
  
import os
if os.path.exists("combined.csv"):
  os.remove("combined.csv")
    
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
dfres =pd.DataFrame(columns = ['test', 'algorithm', 'less', 'my_compare', 'slow_compare', 'sorted', 'stable', 'compare_count'])
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f, sep=';', header=None)
    df.columns = ['test', 'algorithm', 'less', 'my_compare', 'slow_compare', 'sorted', 'stable', 'compare_count']
    # print the location and filename
    #print('Location:', f)
    #print('File Name:', f.split("\\")[-1])
      
    # print the content
    #print('Content:')
    #display(df)
    #pd.concat([dfres, df], axis=1)
    dfres = pd.concat([dfres, df], ignore_index=True)

    #print()
    
dfgrouped = dfres.groupby(['test','algorithm'], 
                  as_index=False).mean()
display(dfgrouped)
dfgrouped.to_csv('combined.csv', sep=';', index=False)
```
